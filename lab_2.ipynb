```python
from math import sin
import matplotlib.pyplot as plt
```

```python
# initial function
def f(x):
    return 13 * sin(x - 13)
```

```python
# tabulation function
def tabulation(a, b, N):
    xl, yl = [], []
    h = (b - a)/N
    x = a
    while a <= x <= b:
        xl.append(x)
        yl.append(f(x))
        x += h
    return [xl, yl]
```

```python
# plot for the initial function
k = 13
N = 26
[x, y] = tabulation(k - 10, k + 10, N)
plt.plot(x, y)
plt.legend(['f(x)'])
plt.show()
```

```python
def w(k, x, xl):
    p = 1.0
    for i in range(k + 1):
        p = p * (x - xl[i])
    return p
```

```python
def rr(k, xl, yl):
    s = 0.0
    for i in range(k + 1):
        p = 1.0
        for j in range(k + 1):
            if i != j:
                p *= (xl[i] - xl[j])
        s += yl[i]/p
    return s
```

```python
def nn(x, N, xl, yl):
    s = yl[0]
    for i in range(1, N):
        s += w(i - 1, x, xl)*(rr(i, xl, yl))
    return s
```

```python
xn, yn, eps = [], [], []
for i in range(len(x)):
    xn.append(x[i])
    yn.append(nn(x[i], len(x), x, y))
    eps.append(abs(y[i] - yn[i]))
```

```python
# initial and aproximate functions plot
plt.plot(x, y)
plt.plot(xn, yn)
plt.legend(['f(x)', 'fn(x)'])
plt.show()
```

```python
# accuracy
plt.plot(x, eps)
plt.legend(['mistake'])
plt.show()
```
