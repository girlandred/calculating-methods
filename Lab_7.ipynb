```python
from math import cos, sin, pi
import matplotlib.pyplot as plt
from random import randint
```

```python
k = 13
N = 100
X = [0]*N
X_answer = [17]*N
A = [ [0]*N for i in range(N) ]

for i in range(N):
    for j in range(N):
        A[i][j] = randint(1, 5)
        
print(A, sep='\n')
```

```python
B = [0]*N

for i in range(N):
    for j in range(N):
        B[i] += A[i][j]*X_answer[j]
```

```python
L = [ [0]*N for i in range(N) ]
U = [ [0]*N for i in range(N) ]

for i in range(N):
    U[i][i] = 1

print(U, sep='\n')
```

```python
def LU_find(L, U, A, n):
    for k in range(n):
        
        for i in range(k, n):
            sum = 0;
            for j in range(k):
                sum += L[i][j]*U[j][k]
            L[i][k] = A[i][k] - sum
            
        for i in range(k + 1, n):
            sum = 0;
            for j in range(k):
                sum += L[k][j]*U[j][i]
            U[k][i] = (A[k][i] - sum)/L[k][k]
```

```python
def LU_solve(L, U, B, X, n):
    Z = [0]*n
    Z[0] = B[0]/L[0][0]
    
    for k in range(1, n):
        sum = 0;
        for j in range(k):
            sum += L[k][j]*Z[j]
        Z[k] = (B[k] - sum)/L[k][k]
        
    X[n - 1] = Z[n - 1]
    
    for k in range(n - 1, -1, -1):
        sum = 0
        for j in range(k, n):
            sum += U[k][j]*X[j]
        X[k] = Z[k] - sum
```

```python
def multiplicationMV(M, V, N):
    R = [0]*N
    for i in range(N):
        sum = 0
        for j in range(N):
            sum += M[i][j] * V[j]
        R[i] = sum
    return R
```
