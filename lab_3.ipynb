```python
from math import sin
import matplotlib.pyplot as plt
```

```python
# initial function
def f(x):
    return 13 * sin(x + 13)
```

```python
# tabulation function
def tabulation(a, b, N):
    xl, yl = [], []
    h = (b - a)/N
    x = a
    while a <= x <= b:
        xl.append(x)
        yl.append(f(x))
        x += h
    return xl, yl
```

```python
# factorial
def factorial(n):
    return 1 if n <= 1 else n * factorial(n - 1)
```

```python
# permutations
def C(n, k):
    return factorial(n) / (factorial(k)*factorial(n - k))
```

```python
# delta f
def df(yl, n, t):
    r = 0
    for k in range(n + 1):
        r += pow(-1, n - k)*C(n, k)*yl[t + k]
    return r
```

```python
# значення факторіального многочлена t(k)
def tk(t, k):
    mn = 1
    if k == 0:
        mn = 1
    else:
        for i in range(k):
            mn *= (t - i)

    return mn
```

```python
# aproximate value for function f(t)
def ft(yl, n, t):
    r = 0
    for k in range(n):
        r += df(yl, k, 0)*tk(t, k)/factorial(k)
    return r
```

```python
N = 30
k = 13

x, y = tabulation(-k, k, N)
plt.plot(x, y)
plt.legend(['f(x)'])
plt.show()
```

```python
xn, yn, eps = [], [], []
for i in range(len(x)):
    xn.append(x[i])
    yn.append(ft(y, len(x), i))
    eps.append(abs(y[i] - yn[i]))
```

```python
plt.plot(x, y)
plt.plot(xn, yn)
plt.legend(['f(x)', 'fn(x)'])
plt.show()
```
